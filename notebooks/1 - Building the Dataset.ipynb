### Getting the First Page

```python
import requests
```

```python
# bulk mode cuts out extraneous data, max per page is 1000
first_page_request = "https://www.speedrun.com/api/v1/games?_bulk=yes&max=1000"

first_page = requests.get(first_page_request)
```

```python
first_page.json()
```
